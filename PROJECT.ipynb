The goal of this project is to perform exploratory data analysis on a modified version of the redwine dataset.The  Red Wine Quality Analysis dataset contains 12 input variables, including fixed acidity, volatile acidity, citric acid, residual sugar, chlorides, free sulfur dioxide, total sulfur dioxide, density, pH, sulphates, and alcohol, with the output variable being the quality score between 0 and 10. ### Description of wine data.

### Cleaning the wineDataset.

```python
wineData.isnull().sum()
```

```python
mean_fixed_acidity = Statistics.fixed_acidity.loc['mean']
mean_volatile_acidity = Statistics.volatile_acidity.loc['mean']
mean_chlorides = Statistics.chlorides.loc['mean']
mean_free_sulfur_dioxide = Statistics.free_sulfur_dioxide.loc['mean']
mean_density = Statistics.density.loc['mean']
mean_quality = Statistics.quality.loc['mean']
```

```python
cleanedWineData = wineData.copy()
```

```python
cleanedWineData.fixed_acidity = cleanedWineData.fixed_acidity.fillna(mean_fixed_acidity)
cleanedWineData.volatile_acidity = cleanedWineData.volatile_acidity.fillna(mean_volatile_acidity)
cleanedWineData.chlorides = cleanedWineData.chlorides.fillna(mean_chlorides)
cleanedWineData.free_sulfur_dioxide = cleanedWineData.free_sulfur_dioxide.fillna(mean_free_sulfur_dioxide)
cleanedWineData.density =cleanedWineData.density.fillna(mean_density)
cleanedWineData.quality = cleanedWineData.quality.fillna(mean_quality)
```

## PIE CHART
The sectors of this pie chart are the sulphates,alcohol,free sulphur dioxide,citric acid,chlorides and average amount of residual sugar from the wineDataSet.

```python
ResidualSugar     = cleanedWineData_statistics['residual_sugar'].loc['mean']
Sulphates         = cleanedWineData_statistics['sulphates'].loc['mean']
Alcohol           = cleanedWineData_statistics['alcohol'].loc['mean']
FreeSulfurDioxide = cleanedWineData_statistics['free_sulfur_dioxide'].loc['mean']
CitricAcid        = cleanedWineData_statistics['citric_acid'].loc['mean']
Chlorides         = cleanedWineData_statistics['chlorides'].loc['mean']

labels = ['Residual sugar', 'Sulphates', 'Alcohol', 'Free sulfur dioxide', 'Citric acid', 'Chlorides']
Values = [0,0,0,0,0,0.6]

mat.pie([ResidualSugar, Sulphates, Alcohol, FreeSulfurDioxide, CitricAcid, Chlorides],
labels = labels, autopct = '%.2f %%', explode = Values)
mat.title('THE COMPONENT OF THE RED WINE', fontdict = {'fontweight' : 'bold'}, ha = 'center')
mat.show()
```

<b>INTERPRETING THE PIE CHART ABOVE <b> <br>
The analysis of red wine quality utilized the mean quality of its various components to simplify the comparison, treating the components as if they belong to a single red wine. This approach was adopted to evaluate the percentage composition in red wine from an aggregated perspective. The goal was to assess the red wine's quality based on this collective view, despite the detailed individual qualities of different red wines. This method aimed to streamline the comparison of red wines by considering their components as a unified entity.
